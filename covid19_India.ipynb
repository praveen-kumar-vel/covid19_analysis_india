```python
import pandas as pd 
import matplotlib.pyplot as plt
data = pd.read_json("https://api.covid19india.org/state_district_wise.json")
fig, ax = plt.subplots()
df = pd.DataFrame(data)
df.columns
states =list(df.columns)
state_list = []
count_list = []
for co in states:
    state_individual = pd.DataFrame(df[co][0])
    count = state_individual.loc["confirmed"]
    #print(co,sum(count))
    count_list.append(sum(count))
    state_list.append(co)  
plt.rcParams["figure.figsize"] = (20,15)
ax.barh(state_list,count_list, align='center',color='red')
ax.set_xlabel('Count ')
ax.set_title('COVID 19 ')
```

```python
table = pd.DataFrame(count_list,state_list)
table1=table.sort_values(by=0).to_dict()
tn = table1[0]
print("Total Count in Tamil Nadu : ",tn["Tamil Nadu"])
```

```python
table_all_state = pd.DataFrame(count_list,state_list)
table_all_state_sorted=table_all_state.sort_values(by=0,ascending=False)
table_all_state_sorted
```

```python
#print(state_list)
#count_list

datafive = pd.DataFrame(count_list,index=state_list)
datafive_sort = datafive.sort_values(by=0)
datafive_tail = datafive_sort.tail()
color = ["#ff9999","#ff4d4d","#ff1a1a","#e60000","#990000",] 

plt.bar(datafive_tail.index,datafive_tail[0],color=color)
plt.rcParams["figure.figsize"] = (20,15)
plt.show()
```

```python
dataf1 = [state_list,count_list]
dataf = pd.DataFrame(dataf1)
data1 = dataf.T
piedata = data1.sort_values(by=1).tail(10)
state=list(piedata[0])
counts = list(piedata[1])

labels = state
sizes = counts
explode = (0, 0, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Top 10 State")


plt.show()
```

```python
dataf1 = [state_list,count_list]
dataf = pd.DataFrame(dataf1)
data1 = dataf.T
piedata = data1.sort_values(by=1).tail()
color = ["#ff9999","#ff4d4d","#ff1a1a","#e60000","#990000",] 
state=list(piedata[0])
counts = list(piedata[1])

labels = state
sizes = counts
explode = (0, 0, 0, 0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=explode, colors=color, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Top 5 State")


plt.show()
```
